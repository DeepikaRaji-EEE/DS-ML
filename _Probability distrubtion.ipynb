```python
from scipy.stats import norm
# Given values
mu = 150  # mean
sigma = 10  # standard deviation
X = 160  # value to check
prob_greater = 1 - norm.cdf(X, mu, sigma)
print("Probability tree is taller than 160 cm:", round(prob_greater * 110, 4), "%")
```

```python

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
mu = 160
sigma = 30
X = np.random.normal(mu, sigma, 10000)
plt.hist(X, bins=40, density=True, alpha=0.5)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 1000)
y = norm.pdf(x, mu, sigma)
plt.plot(x, y, linewidth=1.5)
plt.title("Normal Distribution Histogram\nBell Shaped Curve")
plt.xlabel("Height (cm)")
plt.ylabel("Density")
plt.grid(True)
plt.show()
```

```python
from scipy.stats import expon
import numpy as np
import matplotlib.pyplot as plt

# Given values
mean_time = 10  
t = 5  

lambda_value = 1 / mean_time
prob_within = expon.cdf(t, scale=1/lambda_value)

print("Probability customer arrives within 5 minutes:",
      round(prob_within * 100, 2), "%")
```

```python
import numpy as np
import matplotlib.pyplot as plt

mean_time = 10  # average customer arrival time

# generate random exponential data
data = np.random.exponential(mean_time, 10000)

# histogram
plt.hist(data, bins=30, density=True)  

plt.title("Exponential Distribution - Histogram")
plt.xlabel("Time (minutes)")
plt.ylabel("Density")
plt.grid(True)
plt.show()
```

```python
from scipy.stats import uniform
import numpy as np
import matplotlib.pyplot as plt
# Uniform Distribution range
a = 1      # minimum
b = 110    # maximum
# Interval to check
low = 10
high = 50
dist = uniform(loc=a, scale=b-a)
probability = dist.cdf(high) - dist.cdf(low)
```
