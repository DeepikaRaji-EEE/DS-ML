```python
import numpy as np

# simulated real-time data (e.g., latency in ms)
data = np.random.normal(200, 30, 1000)

n = len(data)
mean = np.mean(data)
std = np.std(data)

# 95% confidence interval
z = 1.96

margin = z * (std / np.sqrt(n))

lower = mean - margin
upper = mean + margin

print("Mean:", mean)
print("95% CI:", (lower, upper))
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.normal(200, 30, 100)

rolling_mean = np.cumsum(data) / np.arange(1, len(data)+1)
rolling_std = [np.std(data[:i]) if i > 1 else 0 for i in range(1, len(data)+1)]

n = np.arange(1, len(data)+1)
margin = 1.96 * np.array(rolling_std) / np.sqrt(n)

upper = rolling_mean + margin
lower = rolling_mean - margin

plt.plot(rolling_mean, label="Mean", color="blue")
plt.fill_between(range(len(data)), lower, upper, color="lightblue", alpha=0.5, label="95% CI")

plt.title("Real-Time Confidence Interval")
plt.legend()
plt.show()
```
