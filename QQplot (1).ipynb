```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# sample data
data = [15, 42, 33, 65, 26, 28, 53, 62]

# Q-Q plot
stats.probplot(data, dist="norm", plot=plt)

plt.title("Normal Q-Q Plot")
plt.show()
```
