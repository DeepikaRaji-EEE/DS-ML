```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

lam = 38
k = 25

# probability
prob = poisson.pmf(k, lam)

print("Probability of getting 15 calls in a minute:", prob)

# distribution
x = np.arange(0, 21)
fit = poisson.pmf(x, lam)

plt.bar(x, fit, color="lightgreen", edgecolor="black")
plt.title("Poisson Distribution")
plt.xlabel("Number of calls per minute")
plt.ylabel("Probability")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

lam = 4
k = 8

# PMF values
x = np.arange(0, 12)
pmf_values = poisson.pmf(x, lam)

result = poisson.pmf(k, lam)

# Plot full distribution
plt.bar(x, pmf_values,
        color="lightblue",
        edgecolor="black",
        label="Poisson PMF")

# Highlight k
plt.bar(k, result,
        color="green",
        edgecolor="black",
        label=f"k = {k}")

# labels
plt.xlabel("Number of arrivals per hour")
plt.ylabel("Probability")
plt.title("Poisson Distribution (λ = 2)")

plt.legend()
plt.grid(axis="y", alpha=0.5)

plt.show()
```
