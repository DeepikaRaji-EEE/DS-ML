```python
import matplotlib.pyplot as plt

# generate data
data = np.random.normal(0, 2, 1000)

# plot graph
plt.hist(data, bins=20)
plt.title("Normal Distribution")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# generate data
data = np.random.uniform(0, 4, 1000)

# plot graph
plt.hist(data, bins=40)
plt.title("Uniform Distribution")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# generate data
data = np.random.exponential(10, 1000)

# plot graph
plt.hist(data, bins=20)
plt.title("Exponential Distribution")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```
